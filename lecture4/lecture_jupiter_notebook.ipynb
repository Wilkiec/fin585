**Larger Datasets: A Quick Look at Speed**

+ Certain methods or functions are a little slow in `Pandas` for "biggish" datasets. 

+ The **BYU Finance library** can be helpful for some bread and butter data analysis tasks because it's functions are more focused speed.

+ Also, when working with lots of data you might like progress bars.

+ Lets look at daily stock market data $\rightarrow$ over 100 million observations.

+ The data is in the `feather` format. This leads to a nice load time improvement for "biggish" data over `CSV` files.

```python
import pandas as pd
import numpy as np
from finance_byu.rolling import roll_std
```

```python
dstk = pd.read_feather('dstk.ftr')[['permno','caldt','ret']]    
dstk
```

```python
dstk.groupby('permno')['ret'].transform(roll_std,250,125,ddof=0)
```

```python
from tqdm.auto import tqdm
tqdm.pandas()

dstk.groupby('permno')['ret'].progress_transform(roll_std,win=250,minp=125,ddof=0)
```

```python
%timeit -n 1 -r 1 dstk.groupby('permno')['ret'].rolling(250,125).std(ddof=0)
```

```python
%timeit -n 1 -r 1 dstk.groupby('permno')['ret'].transform(roll_std,250,125,ddof=0)
```

```python
from lib.dslutils import roll_std
```

```python
std = lambda x: roll_std(x.values,win=250,minp=125)
%timeit -n 1 -r 1 dstk.groupby('permno')['ret'].transform(std)
```
