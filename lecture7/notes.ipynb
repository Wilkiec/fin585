The Sharpe-Litner CAPM

assumptions:
stock returns are jointly multivariate normal

investors like expected return but are averse to risk

securities are infinitely divisible

unrestricted and costless short-selling

riskfree borrowing and lending

complete agreement -- workhorse assumption for CAPM... we agree on what the expected return a vector is and the expected covariance and variance.
